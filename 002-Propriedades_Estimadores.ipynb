## Propriedades dos Estimadores

Seja $\theta$ um parâmetro desconhecido de uma população, como a média, variância ou proporção. Um **estimador** $\hat{\theta}$ é uma função dos dados de uma amostra aleatória que é utilizada para fornecer uma estimativa do valor de $\theta$. A amostra pode consistir de variáveis contínuas, discretas, ou até mesmo atributos categóricos, como variáveis binárias.

Considere um exemplo simples: a renda familiar. A renda familiar é uma variável contínua, e uma característica importante dessa variável é a sua média, que pode ser um parâmetro a ser estimado. Temos à disposição as informações da amostra: $(y_{1}, y_{2}, \dots, y_{n})$. Como $\hat{\theta}$ é uma função de valores amostrais, podemos escrever:

$$
\hat{\theta} = \hat{\theta}(y_{1}, y_{2}, \dots, y_{n})
$$

A única restrição é que a função de valores da amostra, ou seja, a equação do estimador, não pode ter parâmetros desconhecidos. Assim, **o estimador é uma função de variáveis aleatórias, logo, também é uma variável aleatória**.

Isso significa que, **enquanto $\theta$ é um valor fixo (mas desconhecido), $\hat{\theta}$ tem uma distribuição porque depende dos dados amostrais que são aleatórios**.

### Características Básicas de um Estimador $\hat{\theta}$

* Medidas de Tendência Central e Dispersão:
  - **Média:** $E(\hat{\theta})$ 
    - A esperança ou valor esperado de $\hat{\theta}$, que nos diz o valor médio que o estimador assume em diferentes amostras.
  - **Variância:** $\text{Var}(\hat{\theta}) = E[(\hat{\theta} - E(\hat{\theta}))^2]$ 
    - A variabilidade ou dispersão dos valores que $\hat{\theta}$ pode assumir ao redor da sua média.
  - **Desvio Padrão (S.D.):** $\sqrt{\text{Var}(\hat{\theta})}$ 
    - A raiz quadrada da variância de $\hat{\theta}$, indicando a dispersão dos valores do estimador em torno de sua média.

* Medidas de Precisão:
  - **Erro Padrão (S.E.):** $\sqrt{\hat{\text{Var}(\hat{\theta})}}$ 
    - O erro padrão do estimador é a estimativa do desvio padrão da distribuição amostral de $\hat{\theta}$ e indica a precisão da estimativa.
  - **Erro Padrão da Média (S.E. da Média):** $\frac{\sigma}{\sqrt{n}}$
    - Quando estamos lidando especificamente com a média amostral, o erro padrão é dado pela raiz quadrada da variância dividida pelo tamanho da amostra \( n \). Isso reflete como a precisão da média melhora com o aumento do tamanho da amostra.
  - **Erro Amostral:** $Y - \theta$ 
    - A diferença entre o valor observado em uma amostra $Y$ e o valor verdadeiro do parâmetro $\theta$.
  - **Erro Quadrático Médio (MSE):** $E[(\hat{\theta} - \theta)^2]$ 
    - Mede a média dos quadrados das diferenças entre as estimativas e o valor verdadeiro do parâmetro. Combina tanto a variância quanto o viés do estimador.

* Medida de Viés:
  - **Viés:** $E(\hat{\theta}) - \theta$ 
    - A diferença entre a média do estimador e o verdadeiro valor do parâmetro. Um estimador é não viesado se o viés for zero.




Repare que o Erro Amostral (SE), o Erro Quadrático Médio (MSE) e o Viés **dependem do verdadeiro valor de $\theta$, que é desconhecido**. Em práticas estatísticas, essas quantidades são frequentemente estimadas com base nos dados disponíveis e nos estimadores utilizados. Como $\theta$ é desconhecido, os resultados e **análises que envolvem esses termos geralmente assumem que $\theta$ é substituído pelo seu estimador $\hat{\theta}$ para realizar inferências e estimativas práticas** sobre o parâmetro desconhecido.

O MSE mede a dispersão entre o estimador $\hat{\theta}$ e o verdadeiro valor do parâmetro $\theta$. 

* Se o viés for nulo, o estimador não é viesado, então o MSE é igual à variância. 

Para melhor entender essa afirmação, vamos decompor o MSE:

$$ \text{MSE}(\hat{\theta}) = E[(\hat{\theta} - E(\hat{\theta}) + E(\hat{\theta}) - \theta)^2] $$

Expandindo o quadrado:

$$ \text{MSE}(\hat{\theta}) =  E[(\hat{\theta} - E(\hat{\theta}) + E(\hat{\theta}) - \theta)*(\hat{\theta} - E(\hat{\theta}) + E(\hat{\theta}) - \theta)] $$

$$ \text{MSE}(\hat{\theta}) =  E[(\hat{\theta} - E(\hat{\theta}))^2] + E[(E(\hat{\theta}) - \theta)^2] + 2E[(\hat{\theta} - E(\hat{\theta}))(E(\hat{\theta}) - \theta)] $$


O último termo é nulo, pois $ E(\hat{\theta} - E(\hat{\theta})) = 0 $, o que significa que:

$$ 2E[(\hat{\theta} - E(\hat{\theta})) \cdot (E(\hat{\theta}) - \theta)] = 0 $$

Portanto:

$$ \text{MSE}(\hat{\theta}) = E[(\hat{\theta} - E(\hat{\theta}))^2] + (E(\hat{\theta}) - \theta)^2 $$

Ou seja, o valor do MSE nunca será menor do que a variância. O MSE será igual à variância quando o estimador for não viesado.

Compreender essas propriedades básicas do estimador é fundamental para avaliar a qualidade de diferentes estimadores. Em práticas de modelagem econométrica, a minimização do MSE é frequentemente utilizada como um critério para a seleção de estimadores. A seguir, exploraremos mais detalhadamente as características que fazem de um estimador um bom estimador, e como essas propriedades se manifestam tanto em amostras finitas quanto em amostras grandes.



## Propriedades Desejáveis de um Estimador

Normalmente, um bom estimador é aquele que é **não viesado (*Unbiasedness*), consistente (*Consistency*) e eficiente (*Efficiency*)**. E buscamos estimadores que possuam essas propriedades tanto para amostras finitas quanto para amostras grandes. Ou seja:

* **Propriedades de pequenas amostras ou de amostras finitas:** válidas para qualquer tamanho de amostra;
* **Propriedades de grandes amostras, ou assintóticas:** válidas para grandes amostras ($n \to \infty$).

OBS: Em amostras finitas, não-viés e eficiência são propriedades mais críticas para avaliar a qualidade de um estimador.

### Propriedades para Amostras Finitas

* **Não viés (*Unbiasedness*)**

Então, um estimador $ \hat{\theta} $ é não viesado para $ \theta $ se $ E(\hat{\theta}) = \theta $. A média é o valor central da distribuição. Quando se diz que a esperança matemática (E) é igual ao verdadeiro valor do parâmetro, o que se quer afirmar é que a distribuição amostral do estimador $ \hat{\theta} $, $ f(\hat{\theta}) $, gera estimativas de $ \theta $, acima e abaixo de $ \theta $, mas a média desses valores será igual a $ \theta $. Um exemplo de um estimador não viesado é a média amostral como um estimador da média populacional, pois:

$$ E(\bar{Y}) = \frac{1}{n} \sum_{i=1}^{n} Y_i = \mu $$

em que $ \mu $ é a média de $Y$ na população.

* **Eficiência (*Efficiency*)**

No entanto, saber que o estimador produz estimativas cuja média é igual ao verdadeiro valor do parâmetro na população não é suficiente para caracterizá-lo como um bom estimador. Se as estimativas forem muito dispersas em torno do verdadeiro parâmetro, o estimador pode não ser útil. Um estimador que tenha pequena dispersão ou variância, mas seja viesado, é menos útil, especialmente se o viés for desconhecido em magnitude e direção. Por exemplo, um estimador com variância nula, como uma constante, não seria útil. Se escolhemos uma constante como 10, então todos os valores gerados por $ f(\hat{\theta}) $ estariam concentrados em 10, ignorando completamente as informações amostrais. Por isso, seria desejável que um estimador minimizasse o Erro Quadrático Médio (MSE), definido anteriormente. No entanto, o MSE perfeito frequentemente requer o conhecimento do próprio parâmetro $ \theta $, tornando o estimador impraticável. Isso nos leva ao **conceito de eficiência**. **Um estimador é eficiente se for não viesado e tiver variância mínima em comparação a outros estimadores não viesados**.

**$ \hat{\theta} $ é um estimador eficiente de $ \theta $ se**:
1. $ \hat{\theta} $ é não viesado;
2. $ \text{Var}(\hat{\theta}) \leq \text{Var}(\tilde{\theta}) $ para qualquer outro estimador não viesado $ \tilde{\theta} $ de $ \theta $.

Um estimador assim definido é chamado de Melhor Estimador Não Viesado (MVUE). Um estimador viesado não pode ser eficiente, independentemente do viés. O conceito de eficiência pode ser complicado, pois requer a comparação com todos os possíveis estimadores não viesados. Na prática, comparamos a eficiência de um estimador em relação a outro, simplificando o conceito. Em muitas situações, **a eficiência pode ser avaliada em termos absolutos usando o Teorema de Cramer-Rao**, que define um limite inferior para a variância de estimadores não viesados.



#### Teorema de Cramer-Rao

Considere $ Y $ uma variável aleatória com função de probabilidade $ f(Y) $, caracterizada por parâmetros $ \theta_1, \theta_2, ..., \theta_K $. Considere $ \hat{\theta}_1 $ um estimador não viesado de $ \theta_1 $, derivado de uma amostra $ y_1, y_2, ..., y_n $ de $ Y $. Defina $ l = \ln L = \ln f(y_1, y_2, ..., y_n) $, conhecida como função log-verossimilhança (logaritmo neperiano da função de verossimilhança $ L $ da amostra). A matriz Hessiana, composta pelas segundas derivadas da função log-verossimilhança, é dada por:

$$
H_{KK} = \begin{bmatrix}
-E\left[\frac{\partial^2 l}{\partial \theta_1^2}\right] & \ldots & -E\left[\frac{\partial^2 l}{\partial \theta_1 \partial \theta_K}\right] \\
\vdots & \ddots & \vdots \\
-E\left[\frac{\partial^2 l}{\partial \theta_K \partial \theta_1}\right] & \ldots & -E\left[\frac{\partial^2 l}{\partial \theta_K^2}\right]
\end{bmatrix}
$$

Essa matriz é conhecida como matriz de informação de Fisher. Considere agora a inversa da matriz de informação, $ I_{KK} = H_{KK}^{-1} $. 

Então, a desigualdade de Cramer-Rao é definida por:

$$ \text{Var}(\hat{\theta}_i) \geq I_{ii}, \quad i = 1, ..., K $$

onde $ I_{ii} $ é o elemento da diagonal da inversa da matriz de informação. Este teorema nos permite construir um limite inferior (maior que zero) para a variância de qualquer estimador não viesado, desde que conheçamos a função de distribuição da população a partir da qual os dados foram gerados. Esse limite inferior é chamado de Limite Inferior de Cramer-Rao, que nos garante que nenhum estimador não viesado terá variância menor que este limite; o máximo que pode ocorrer é da variância do estimador ser igual ao limite de Cramer-Rao, jamais será inferior. Se isso ocorre, sabemos que nenhum outro estimador não viesado terá variância menor que a deste estimador e podemos afirmar que ele é eficiente, no sentido absoluto do conceito. Sabemos que $ \text{Var}(\bar{Y}) = \frac{\sigma^2}{n} $. Se a distribuição geradora da amostra for Normal, pode-se mostrar que $ \frac{\sigma^2}{n} $ é o limite inferior de Cramer-Rao para estimadores não viesados da média populacional; portanto, a média amostral é um estimador eficiente da média populacional para uma população normal.

É importante observar que nem sempre o limite inferior de Cramer-Rao funciona para um estimador, porque nem sempre ele é atingido por um estimador. Por exemplo, o limite inferior que precisa ser atingido para que o estimador da variância de uma população Normal seja eficiente é dado por $ \frac{2\sigma^4}{n} $, mas não existe nenhum estimador não viesado de $ \sigma^2 $ que tenha variância igual a este limite.

Esta discussão mostra que o conceito de estimador eficiente entre todos os estimadores não é trivial. Para obter um conceito de eficiência de mais fácil aplicação, é melhor que nos restrinjamos a uma classe específica de estimadores. No nosso caso, focaremos naqueles que são os mais frequentemente utilizados em análise empírica em economia: **estimadores que sejam funções lineares das observações da amostra** (estimadores lineares). Isso nos leva a um conceito mais especializado de eficiência.

(OBS: no final dessas notas você encontrará um resumo do procedimento e um exemplo de aplicação do Teorema de Cramer-Rao para um estimador de variância de uma população assumindo uma distribuição Normal.)

### Eficiência para Estimadores Lineares

Um estimador $ \hat{\theta} $ é o melhor estimador linear não viesado (da sigla em inglês: BLUE) de $ \theta $ se as seguintes condições forem satisfeitas:
* $ \hat{\theta} $ é uma **função linear** das observações da amostra;
*  $ \hat{\theta} $ é um estimador **não viesado** de $ \theta $; e
*  $ \text{Var}(\hat{\theta}) \leq \text{Var}(\tilde{\theta}) $, onde $ \tilde{\theta} $ é qualquer outro estimador linear não viesado de $ \theta $.

A condição de linearidade significa que para uma amostra $ y_1, y_2, ..., y_n $, o estimador tem que ter a forma $ a_1y_1 + a_2y_2 + \ldots + a_ny_n $, que é exatamente aquela apresentada, por exemplo, para a média amostral:

$$ \bar{Y} = a_1y_1 + a_2y_2 + \ldots + a_ny_n =  \frac{1}{n}(y_1 + y_2 + \ldots + y_n) $$

em que $ a_1 = a_2 = \ldots = a_n = \frac{1}{n} $.

Ou seja, o estimador da média amostral é linear nas observações da amostra. Como já vimos, ele é também não viesado e sua variância é igual ao limite mínimo de Cramer-Rao quando a distribuição geradora da amostra for normal. Como consequência, a média amostral é um estimador BLUE (Best Linear Unbiased Estimator) da média populacional. Em outras palavras, para ser *Best Linear Unbiased Estimator* (Melhor Estimador Linear Não Viesado) um estimador é linear nas observações da amostra, não viesado e tem a menor variância possível entre todos os estimadores lineares não viesados.


### Propriedades Assintóticas

No contexto de amostras grandes, um bom estimador é aquele que é **não viesado (*Unbiasedness*), consistente (*Consistency*) e eficiente (*Efficiency*)**.

Propriedades assintóticas são aquelas válidas apenas para grandes amostras, ou para amostras com tamanho $n$ se aproximando do infinito. A distribuição amostral de um estimador é diferente para tamanhos de amostras diferentes. **A distribuição da média amostral para amostras de qualquer população é definida pelo Teorema do Limite Central**.

#### Teorema do Limite Central

Para qualquer que seja a distribuição de $Y$, com média $ \mu $ e variância $ \sigma^2 $, a distribuição da média amostral $ \bar{Y} $ converge em distribuição para $ N(\mu, \frac{\sigma^2}{n}) $ (normal com média $ \mu $ e variância $ \frac{\sigma^2}{n} $) quando o tamanho da amostra, $n$, converge para infinito. Portanto, a distribuição da média amostral em grandes amostras se aproxima de uma normal, independentemente da distribuição original de $Y$.

O Teorema nos garante que, quando o tamanho da amostra cresce, a distribuição assintótica da média amostral será aproximadamente normal. Então, independentemente da distribuição que tenha gerado os dados amostrais, sabemos que a distribuição da média amostral convergirá para uma normal, sendo esta a distribuição assintótica do estimador.

Quando falamos em distribuição assintótica, estamos nos referindo à forma que a distribuição de um estimador assume à medida que o tamanho da amostra se torna muito grande. À medida que $n$ aumenta, a variância da distribuição da média amostral diminui, o que faz com que a distribuição se torne cada vez mais concentrada em torno do verdadeiro valor do parâmetro que estamos tentando estimar. Em termos técnicos, à medida que $n$ aumenta, a variância da distribuição do estimador diminui, o que faz com que a distribuição "colapse" ou se concentre em torno do verdadeiro valor do parâmetro. Uma distribuição que se concentra totalmente em um único ponto é chamada de degenerada. Essa ideia de concentração ou degeneração ajuda a entender por que, em amostras muito grandes, o estimador fornece uma estimativa cada vez mais precisa do verdadeiro valor do parâmetro. Em outras palavras, o estimador se torna mais confiável à medida que a amostra cresce.

Vamos usar a média amostral como exemplo. Sabemos que, para qualquer tamanho da amostra:

$$ E(\bar{Y}) = \mu $$

em que $ \mu $ é a média de $Y$ na população; e

$$ \text{Var}(\bar{Y}) = \frac{\sigma^2}{n} $$

em que $ \sigma^2 $ é a variância de $Y$ na população e $n$ é o tamanho da amostra. Neste caso, é claro que a distribuição da média amostral se torna cada vez mais estreita em torno da média populacional conforme $n$ aumenta. A distribuição se concentra quando $n$ vai para infinito, pois a variância se torna cada vez menor (conforme $n$ cresce, a distribuição vai ficando cada vez mais estreita, com a diminuição da variância à medida que a amostra cresce). No entanto, mesmo que a distribuição que gerou os dados não seja normal, a distribuição da média amostral se aproxima de uma normal, conforme garantido pelo Teorema do Limite Central. Esta é a distribuição antes do colapso final.

Tendo discutido a distribuição assintótica, vamos verificar as possíveis formas que ela pode assumir. Basicamente, três formas englobam todas as possibilidades:

* A distribuição assintótica do estimador é conhecida e é a mesma para qualquer tamanho da amostra. Um exemplo é a média amostral como estimador da média de uma população normal. Neste caso, **a distribuição da média amostral é normal para qualquer tamanho da amostra**, com média $ \mu $ e variância $ \frac{\sigma^2}{n} $.
* **A distribuição do estimador não é da mesma forma para diferentes tamanhos de amostra**. Este é o caso da distribuição da proporção de sucessos, que tem distribuição binomial, mas tem distribuição assintótica normal, como garantido pelo TLC.
* Por vezes, a forma da distribuição do estimador é desconhecida para diferentes tamanhos da amostra, exceto para a distribuição assintótica (com o tamanho da amostra convergindo para o infinito). É o caso da média amostral de uma população não normal. Podemos desconhecer a sua distribuição para amostras finitas, mas sabemos que a distribuição assintótica é normal, garantida pelo TLC.

Essas três categorias de estimadores cobrem praticamente todos os problemas de estimação em econometria. De forma similar à distribuição para amostras finitas, as distribuições assintóticas são também caracterizadas por dois momentos: média e variância assintóticas.

A média da distribuição assintótica do estimador é o valor assumido pela média quando o tamanho da amostra converge para infinito:

**Média Assintótica:** é dada por

$$ \lim_{n \to \infty} E(\hat{\theta}) $$

desde que a média de $ \hat{\theta} $ exista. No entanto, a variância não é dada por

$$ \lim_{n \to \infty} \text{Var}(\hat{\theta}) $$

Para verificar este fato, basta olhar para

$$ \lim_{n \to \infty} \text{Var}(\bar{Y}) = \lim_{n \to \infty} \frac{\sigma^2}{n} = 0 $$

Para essa definição de variância assintótica, temos que a variância quase sempre é nula. Mas o que queremos de fato é a variância da distribuição assintótica, que, por força do

* **Viés Assintótico**

Um estimador é assintoticamente não viesado se:

$$ \lim_{n \to \infty} E(\hat{\theta}) = \theta $$

Por exemplo, considere o estimador da variância de $ Y \sim N(\mu, \sigma^2) $:

$$ \hat{\sigma}^2 = \frac{1}{n} \sum_{i=1}^{n} (Y_i - \bar{Y})^2 $$

Este é um estimador viesado para $ \sigma^2 $ em amostras finitas. Vamos ver por quê.

Se tomarmos a expectativa da variância amostral $\hat{\sigma}^2$, podemos decompor a expressão como segue:

$$ E(\hat{\sigma}^2) = \frac{1}{n} \sum_{i=1}^{n} E\left[(Y_i - \mu)^2 - 2(Y_i - \mu)(\bar{Y} - \mu) + (\bar{Y} - \mu)^2\right] $$

Ao expandir e simplificar, obtemos:

$$ E(\hat{\sigma}^2) = \sigma^2 - \frac{\sigma^2}{n} $$

Portanto, o viés do estimador é dado por:

$$ \text{Viés}(\hat{\sigma}^2) = E(\hat{\sigma}^2) - \sigma^2 = -\frac{\sigma^2}{n} $$

No entanto, à medida que $n$ tende ao infinito, este viés tende a zero:

$$ \lim_{n \to \infty} \text{Viés}(\hat{\sigma}^2) = 0 $$

Logo, $\hat{\sigma}^2$ é um estimador assintoticamente não viesado.

* **Consistência**

Um estimador $\hat{\theta}$ é consistente se ele convergir em probabilidade para o verdadeiro valor do parâmetro $\theta$ à medida que o tamanho da amostra aumenta:

$$ \text{plim}(\hat{\theta}) = \theta $$

Isso significa que para qualquer $\epsilon > 0$:

$$ \lim_{n \to \infty} P(|\hat{\theta} - \theta| < \epsilon) = 1 $$

Em outras palavras, conforme $n$ aumenta, a probabilidade de $\hat{\theta}$ estar arbitrariamente próximo de $\theta$ tende a 1.

Outra forma de entender a consistência é através do Erro Quadrático Médio (MSE). Um estimador é consistente se o MSE tende a zero quando $n$ tende ao infinito:

$$ \lim_{n \to \infty} MSE(\hat{\theta}) = 0 $$

O MSE pode ser decomposto em viés ao quadrado mais a variância:

$$ MSE(\hat{\theta}) = \text{Viés}(\hat{\theta})^2 + \text{Var}(\hat{\theta}) $$

Assim, para que $\hat{\theta}$ seja consistente, tanto o viés quanto a variância devem tender a zero conforme $n$ aumenta.


* **Eficiência Assintótica**

A eficiência assintótica refere-se à comparação da variância assintótica de um estimador com o limite inferior estabelecido pela desigualdade de Cramer-Rao. Um estimador $\hat{\theta}$ é assintoticamente eficiente para estimar $\theta$ se satisfizer as seguintes condições:

1. $\hat{\theta}$ tem distribuição assintótica com média $\theta$ e variância finita;
2. $\hat{\theta}$ é consistente; e
3. Nenhum outro estimador consistente de $\theta$ tem variância assintótica menor que a variância assintótica de $\hat{\theta}$.

Se essas condições forem atendidas, então $\hat{\theta}$ é considerado assintoticamente eficiente. Essa propriedade é particularmente relevante em situações onde estamos interessados em encontrar o estimador com a menor variância possível à medida que o tamanho da amostra cresce.

O estimador de máxima verossimilhança (MLE) é frequentemente assintoticamente eficiente, atingindo o limite de Cramer-Rao, tornando-se uma escolha comum em muitas aplicações estatísticas.

* **Simulação das Médias Amostrais**
  - **Função simulate_means**: 
     - A função `simulate_means` é usada para calcular as médias amostrais. Ela recebe como parâmetros uma função de distribuição (`dist_func`), o tamanho da amostra (`sample_size`), e o número de amostras a serem geradas (`n_samples`).
     - Para cada amostra, a função gera uma amostra aleatória da distribuição definida e calcula sua média. Este processo é repetido 10.000 vezes (definido por `n_samples`), resultando em uma lista de médias amostrais.

* **Distribuição Original: Exponencial**
  - A função `dist_func = lambda size: np.random.exponential(scale=2, size=size)` define a distribuição original dos dados como exponencial, com um parâmetro de escala 2. A distribuição exponencial é assimétrica e positiva, o que torna interessante observar como a média amostral, que inicialmente não é simétrica, se aproxima de uma distribuição simétrica à medida que o tamanho da amostra aumenta.

* **Visualização das Distribuições das Médias Amostrais**
   - **Plotando os Resultados**: 
     - O código usa a biblioteca `seaborn` para gerar histogramas das médias amostrais com um kernel density estimate (KDE) sobreposto, o que facilita a visualização da densidade da distribuição das médias.
     - Para cada tamanho de amostra, é gerado um histograma mostrando como as médias amostrais estão distribuídas. À medida que o tamanho da amostra aumenta, os alunos devem observar que a distribuição das médias se torna mais simétrica e mais próxima de uma normal.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parâmetros da simulação
n_samples = 10000
sample_sizes = [10, 30, 50, 100]

# Função para simular a média amostral
def simulate_means(dist_func, sample_size, n_samples):
    means = [np.mean(dist_func(sample_size)) for _ in range(n_samples)]
    return means

# Distribuição original (por exemplo, exponencial)
dist_func = lambda size: np.random.exponential(scale=2, size=size)

# Plotando a distribuição das médias amostrais para diferentes tamanhos de amostra
plt.figure(figsize=(10, 6))
for size in sample_sizes:
    means = simulate_means(dist_func, size, n_samples)
    sns.histplot(means, kde=True, label=f'n = {size}', stat="density")

plt.title("Distribuição das Médias Amostrais")
plt.xlabel("Média Amostral")
plt.ylabel("Densidade")
plt.legend()
plt.show()
```

* **Interpretação dos Resultados**
  - **n = 10**: Quando o tamanho da amostra é 10, a distribuição das médias amostrais ainda reflete a assimetria da distribuição exponencial original. A forma não é perfeitamente normal, mas já mostra alguma centralização em torno da média.
  - **n = 30**: A distribuição das médias amostrais começa a se aproximar mais de uma forma simétrica, parecendo mais com uma curva normal.
  - **n = 50**: A distribuição é ainda mais simétrica e se aproxima mais da distribuição normal. Isso reflete o início da convergência para a normalidade, como previsto pelo TLC.
  - **n = 100**: Neste ponto, a distribuição das médias amostrais é quase indistinguível de uma distribuição normal. A maioria dos valores está centralizada em torno da média com poucas médias extremas.

### Simulação de Consistência Assintótica

O objetivo desta simulação é demonstrar a propriedade de consistência de um estimador, que é uma característica fundamental em estatística e econometria. A consistência significa que, à medida que o tamanho da amostra aumenta, o estimador converge para o verdadeiro valor do parâmetro que está sendo estimado. Neste exemplo, estamos analisando a consistência da média amostral.

* Configuração dos Parâmetros
  * true_mean = 5: Este parâmetro define o valor real da média da população. Neste caso, estamos assumindo que a verdadeira média da população é 5.
  * sample_sizes = np.arange(10, 1001, 50): Aqui, estamos definindo uma faixa de tamanhos de amostra que varia de 10 a 1000, aumentando em incrementos de 50. Isso permite observar como a média amostral se comporta à medida que o tamanho da amostra cresce.

* Simulação das Médias Amostrais
  * Cálculo da Média Amostral:
    * Para cada tamanho de amostra definido em sample_sizes, o código gera uma amostra aleatória de uma distribuição normal com média true_mean e desvio padrão 2.
    * A média dessa amostra é então calculada e armazenada na lista means.
    * Este processo simula a repetição de experimentos com diferentes tamanhos de amostra e permite visualizar como a média amostral muda à medida que o tamanho da amostra aumenta.

* Visualização dos Resultados
  * Gráfico da Média Amostral vs. Tamanho da Amostra:
    * O gráfico gerado mostra a média amostral em função do tamanho da amostra. Cada ponto no gráfico representa a média calculada para uma amostra de determinado tamanho.
    * Linha Vermelha (Média Verdadeira): Uma linha vermelha pontilhada é desenhada no valor da verdadeira média (5), que serve como referência para visualizar a proximidade das médias amostrais ao valor real à medida que o tamanho da amostra aumenta.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros da simulação
true_mean = 5  # Valor verdadeiro da média populacional
sample_sizes = np.arange(10, 1001, 50)  # Tamanhos de amostra de 10 até 1000, em passos de 50

# Lista para armazenar as médias amostrais
means = []

# Simulando a média amostral para diferentes tamanhos de amostra
for size in sample_sizes:
    # Gera uma amostra aleatória da distribuição normal com média 5 e desvio padrão 2
    sample = np.random.normal(loc=true_mean, scale=2, size=size)
    # Calcula a média da amostra e armazena na lista
    means.append(np.mean(sample))

# Plotando a média amostral em função do tamanho da amostra
plt.figure(figsize=(10, 6))
plt.plot(sample_sizes, means, marker='o')  # Gráfico das médias amostrais
plt.axhline(y=true_mean, color='r', linestyle='--', label='Média Verdadeira')  # Linha da média verdadeira
plt.title("Consistência do Estimador (Média Amostral)")
plt.xlabel("Tamanho da Amostra")
plt.ylabel("Média Amostral")
plt.legend()
plt.show()
```

* Interpretação dos Resultados
  * Tamanhos de Amostra Pequenos:
    * Com tamanhos de amostra menores, as médias amostrais podem variar significativamente em torno da verdadeira média. Isso ocorre porque, com amostras pequenas, a variabilidade aleatória é maior.
  * Tamanhos de Amostra Grandes:
    * À medida que o tamanho da amostra aumenta, as médias amostrais começam a se concentrar em torno do valor real da média (5). Isso demonstra a consistência do estimador, mostrando que a média amostral se aproxima cada vez mais da média verdadeira à medida que a amostra cresce.

A simulação ilustra claramente a definição de consistência: conforme o tamanho da amostra aumenta, a probabilidade de a média amostral estar perto da verdadeira média aumenta. No limite, com uma amostra suficientemente grande, a média amostral se tornará arbitrariamente próxima da verdadeira média.

### Simulação da Eficiência Assintótica

Esta simulação tem como objetivo comparar a eficiência assintótica de dois estimadores da média de uma distribuição normal. A eficiência assintótica é uma propriedade importante que indica se um estimador atinge a menor variância possível (conforme o limite de Cramer-Rao) à medida que o tamanho da amostra cresce. Estimadores eficientes minimizam a variância do erro de estimação, o que é crucial em aplicações práticas de econometria e estatística.

* Configuração dos Parâmetros
  * true_mu = 5 e true_sigma = 2: Estes são os parâmetros da distribuição normal da qual as amostras serão extraídas. A média verdadeira da população é 5, e o desvio padrão é 2.
  * sample_sizes = np.arange(10, 1001, 50): Esta lista define uma sequência de tamanhos de amostra, começando em 10 e indo até 1000, com incrementos de 50. Isso nos permite observar como a variância dos estimadores se comporta à medida que o tamanho da amostra aumenta.
  * n_simulations = 1000: O número de simulações para cada tamanho de amostra. Para cada tamanho de amostra, 1000 amostras são geradas, e as médias amostrais são calculadas para os dois estimadores.
* Definição dos Estimadores
  * estimator_1(sample_size): Este é o estimador da média amostral padrão. Ele simplesmente calcula a média das observações em cada amostra. Espera-se que este estimador seja eficiente, especialmente porque estamos lidando com uma distribuição normal, onde a média amostral é o estimador de máxima verossimilhança (MLE).
  * estimator_2(sample_size): Este é um estimador alternativo, que calcula a média entre a média amostral e a mediana da amostra. Este estimador foi criado para fins de comparação e não é necessariamente otimizado em termos de eficiência. Sua variância é comparada com a do estimador 1 para avaliar qual deles se aproxima mais do limite de Cramer-Rao.
* Simulação das Variâncias
  * Para cada tamanho de amostra, o código executa 1000 simulações para ambos os estimadores. As médias amostrais geradas por cada estimador são usadas para calcular a variância amostral das médias.
  * Cálculo do Limite de Cramer-Rao: O limite de Cramer-Rao é calculado como $\frac{\sigma^2}{n}$, onde $\sigma^2$ é a variância verdadeira da população e $n$ é o tamanho da amostra. Este limite representa a variância mínima teórica que um estimador não viesado pode atingir.
* Visualização dos Resultados
  * Gráfico das Variâncias: O gráfico gerado mostra a variância amostral de ambos os estimadores em função do tamanho da amostra. Ele também exibe o limite de Cramer-Rao como uma linha de referência. Comparando as curvas dos dois estimadores com o limite de Cramer-Rao, podemos avaliar qual estimador é mais eficiente.
    * Variância do Estimador 1 (Média Amostral): Como a média amostral é o estimador de máxima verossimilhança para uma distribuição normal, espera-se que sua variância se aproxime rapidamente do limite de Cramer-Rao, especialmente à medida que o tamanho da amostra aumenta.
    * Variância do Estimador 2 (Média Ajustada): A variância deste estimador pode não se aproximar do limite de Cramer-Rao tão bem quanto a do estimador 1, refletindo sua menor eficiência.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros da simulação
true_mu = 5
true_sigma = 2
sample_sizes = np.arange(10, 1001, 50)
n_simulations = 1000

# Funções para calcular as médias amostrais com os dois estimadores
def estimator_1(sample_size):
    sample = np.random.normal(loc=true_mu, scale=true_sigma, size=sample_size)
    return np.mean(sample)

def estimator_2(sample_size):
    sample = np.random.normal(loc=true_mu, scale=true_sigma, size=sample_size)
    return (np.mean(sample) + np.median(sample)) / 2

# Simulação
variances_1 = []
variances_2 = []
cramer_rao_limits = []

for size in sample_sizes:
    means_1 = [estimator_1(size) for _ in range(n_simulations)]
    means_2 = [estimator_2(size) for _ in range(n_simulations)]
    
    variance_1 = np.var(means_1, ddof=1)
    variance_2 = np.var(means_2, ddof=1)
    cramer_rao = true_sigma**2 / size
    
    variances_1.append(variance_1)
    variances_2.append(variance_2)
    cramer_rao_limits.append(cramer_rao)

# Plotando as variâncias amostrais dos dois estimadores e o limite de Cramer-Rao
plt.figure(figsize=(10, 6))
plt.plot(sample_sizes, variances_1, marker='o', label='Variância Estimador 1 (Média Amostral)')
plt.plot(sample_sizes, variances_2, marker='o', label='Variância Estimador 2 (Média Ajustada)')
plt.plot(sample_sizes, cramer_rao_limits, 'r--', label='Limite de Cramer-Rao')
plt.title("Comparação das Variâncias Amostrais e o Limite de Cramer-Rao")
plt.xlabel("Tamanho da Amostra")
plt.ylabel("Variância")
plt.legend()
plt.show()
```

* Interpretação dos Resultados
  * Estimador 1 (Média Amostral): Se a curva de variância do estimador 1 se aproxima do limite de Cramer-Rao, isso confirma que ele é assintoticamente eficiente. Esta propriedade é esperada, dado que a média amostral é o estimador de máxima verossimilhança para a distribuição normal.
  * Estimador 2 (Média Ajustada): Se a curva de variância do estimador 2 não se aproxima tão bem do limite de Cramer-Rao quanto a do estimador 1, isso indica que ele é menos eficiente. Isso pode ser esperado, dado que este estimador é uma combinação ad hoc da média e da mediana, sem uma base teórica sólida para ser considerado otimizado em termos de variância.

Esta simulação permite aos alunos observar de forma prática como a eficiência assintótica de diferentes estimadores pode ser comparada. Ela ilustra a importância de escolher estimadores que minimizem a variância, especialmente em amostras grandes, onde a eficiência assintótica se torna um critério crucial.

### Extra

#### Procedimento geral para calcular o limite de Cramer-Rao

Em resumo, para calcular o limite de Cramer-Rao, precisamos seguir os seguintes passos:

1. **Definir a função de verossimilhança:**
   Para uma amostra $ y_1, y_2, \dots, y_n $, a função de verossimilhança é:
   $$
   L(\theta) = f(y_1; \theta) \times f(y_2; \theta) \times \dots \times f(y_n; \theta)
   $$
   onde $ f(y_i; \theta) $ é a função de densidade de probabilidade associada à distribuição dos $ y_i $.

2. **Calcular o logaritmo da função de verossimilhança:**
   O logaritmo da função de verossimilhança é dado por:
   $$
   l(\theta) = \sum_{i=1}^n \ln f(y_i; \theta)
   $$

3. **Derivar a primeira e a segunda derivadas do log-verossimilhança:**
   Calcule a primeira derivada (score) $ \frac{\partial l(\theta)}{\partial \theta} $ e a segunda derivada $ \frac{\partial^2 l(\theta)}{\partial \theta^2} $.

4. **Obter a informação de Fisher:**
   A informação de Fisher é:
   $$
   I(\theta) = -E\left[\frac{\partial^2 l(\theta)}{\partial \theta^2}\right]
   $$

5. **Calcular o limite inferior de Cramer-Rao:**
   O limite inferior para a variância de um estimador não viesado é:
   $$
   \text{Var}(\hat{\theta}) \geq \frac{1}{I(\theta)}
   $$


#### Exemplo de Cálculo do Limite Inferior de Cramer-Rao para a média de uma distribuição Normal

Vamos aplicar esses passos a um exemplo simples, como a média de uma distribuição normal $ Y \sim N(\mu, \sigma^2) $. Para uma amostra $ y_1, y_2, \dots, y_n $ de uma distribuição normal $ Y \sim N(\mu, \sigma^2) $, a função de verossimilhança é:

$$
L(\mu) = \prod_{i=1}^n \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(y_i - \mu)^2}{2\sigma^2}\right)
$$


Tomando o logaritmo:

$$
l(\mu) = \sum_{i=1}^n \left(-\frac{1}{2} \ln(2\pi\sigma^2) - \frac{(y_i - \mu)^2}{2\sigma^2}\right)
$$


A primeira derivada em relação a $ \mu $:

$$
\frac{\partial l(\mu)}{\partial \mu} = \frac{1}{\sigma^2} \sum_{i=1}^n (y_i - \mu)
$$

A segunda derivada:

$$
\frac{\partial^2 l(\mu)}{\partial \mu^2} = -\frac{n}{\sigma^2}
$$

A informação de Fisher é:

$$
I(\mu) = -E\left[\frac{\partial^2 l(\mu)}{\partial \mu^2}\right] = \frac{n}{\sigma^2}
$$


O limite inferior de Cramer-Rao para a variância de $ \hat{\mu} $ é:

$$
\text{Var}(\hat{\mu}) \geq \frac{1}{I(\mu)} = \frac{\sigma^2}{n}
$$

Isso significa que, para um estimador não viesado da média $ \mu $, como a média amostral $ \bar{Y} $, a variância mínima possível é $ \frac{\sigma^2}{n} $.
